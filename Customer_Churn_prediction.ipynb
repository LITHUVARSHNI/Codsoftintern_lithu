REMOVING THE SPECIFIC COLUMNS

```python
data=data.drop(['RowNumber', 'CustomerId', 'Surname'],axis=1)
```

UNIQUE VALUES OF THE COLUMN

```python
data['Geography'].unique()
```

CONVERSION OF CATEGORICAL VALUES TO INDICATOR VARIABLES

```python
data = pd.get_dummies(data,drop_first=True)
```

```python
data.head()
```

COUNT THE OCCURRENCE OF EACH UNIQUE VALUE

```python
data['Exited'].value_counts()
```

```python
X=data.drop('Exited',axis=1)
Y=data['Exited']
```

```python
Y
```

 SMOTE (Synthetic Minority Over-sampling Technique) addresses class imbalance by generating synthetic samples for the minority clas

```python
from imblearn.over_sampling import SMOTE
X_res,Y_res = SMOTE().fit_resample(X,Y)
Y_res.value_counts()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_res, Y_res, test_size=0.2, random_state=42)
```

```python
sc=StandardScaler()
```

```python
X_train =sc.fit_transform(X_train)
X_test =sc.transform(X_test)
```

PERFORM LOGISTIC REGRESSION

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

PERFORM RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,Y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred)
```

```python
from sklearn.metrics import precision_score
precision_score(Y_test,y_pred)
```

SAVING THE TRAINED MODEL

```python
with open('model_pickle','wb') as f:
  pickle.dump(rf,f)
```
